```python
import pandas as pd
import pandasql

from pandasql import sqldf
sql_run = lambda q: sqldf(q, globals())
```

```python
worldcity = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Day 18/worldcities.csv', sep = ';', encoding = 'latin-1')
uscity = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Day 18/uscities.csv', sep = ';', encoding = 'latin-1')
uscity.columns = ['city', 'state_id', 'state_name', 'county_fips', 'county_name',
       'lat', 'lng', 'population', 'density', 'source', 'military',
       'incorporated', 'timezone', 'ranking', 'zips', 'id']
```

### Tutorial Bagaimana Cara Melakukan Query Menggunakan Pandasql

Contoh: Bagaimana melakukan query untuk mendapatkan kota-kota dari worldcity yang merupakan:
- berada di antara 33 - 40 'latitude'
- berada di antara 100 - 140 'longitude'
- merupakan kota 'primary' pada kolom 'capital'
- memiliki populasi lebih dari 20 juta

```python
worldcity.head()
```

```python
query = """
SELECT DISTINCT city FROM worldcity WHERE lat BETWEEN 33 AND 40 AND lng BETWEEN 100 AND 140 AND capital = "primary" AND population > 20000000
"""

sql_run(query)
```

```python
query = """
SELECT DISTINCT city
FROM worldcity as w
JOIN uscity as u on w.id=u.id
"""

sql_run(query)
```
